```python
# Dependencies and Setup
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
from matplotlib.pyplot import figure

# Hide warning messages in notebook
import warnings
warnings.filterwarnings('ignore')

# File to Load (Remember to Change These)
mouse_drug_data_to_load = "data/mouse_drug_data.csv"
clinical_trial_data_to_load = "data/clinicaltrial_data.csv"

# Read the Mouse and Drug Data and the Clinical Trial Data
mouse_drug_data_df = pd.read_csv(mouse_drug_data_to_load)
clinical_trial_data_df = pd.read_csv(clinical_trial_data_to_load)

# Combine the data into a single dataset
merge_data_df = pd.merge(mouse_drug_data_df, clinical_trial_data_df, on = "Mouse ID")

# Display the data table for preview
merge_data_df.head()
```

## Tumor Response to Treatment

```python
# Store the Mean Tumor Volume Data Grouped by Drug and Timepoint 
grouped_df = merge_data_df.groupby(['Drug', 'Timepoint'])
mean_df = grouped_df['Tumor Volume (mm3)'].mean()

# Convert to DataFrame
mean_df = pd.DataFrame(mean_df)
mean_df
# Preview DataFrame
mean_df.head()
```

```python
from statistics import stdev
# Store the Standard Error of Tumor Volumes Grouped by Drug and Timepoint
grouped_df_2 = merge_data_df.groupby(['Drug','Timepoint'])
sterr_df = grouped_df_2.sem()['Tumor Volume (mm3)']
# Convert to DataFrame
sterr_df = pd.DataFrame(sterr_df)

# Preview DataFrame
sterr_df.head()
```

```python
# Minor Data Munging to Re-Format the Data Frames
#mean_df.head()
pivot_mean_df = mean_df.pivot_table(index="Timepoint",columns="Drug")["Tumor Volume (mm3)"]
pivot_mean_df


# pivot_mean_df
# Preview that Reformatting worked
```

```python
pivot_sterr_df = sterr_df.pivot_table(index="Timepoint",columns="Drug")["Tumor Volume (mm3)"]
pivot_sterr_df
```

```python
# Generate the Plot (with Error Bars)
plt.errorbar(pivot_mean_df.index,pivot_mean_df["Capomulin"],yerr=pivot_sterr_df["Capomulin"],color="red",marker="o",linewidth=0.5,elinewidth=1.5,label="Capomulin")
plt.errorbar(pivot_mean_df.index,pivot_mean_df["Infubinol"],yerr=pivot_sterr_df["Infubinol"],color="blue",marker="^",linewidth=0.5,elinewidth=1.5,label="Infubinol")
plt.errorbar(pivot_mean_df.index,pivot_mean_df["Ketapril"],yerr=pivot_sterr_df["Ketapril"],color="green",marker="s",linewidth=0.5,elinewidth=1.5,label="Ketapril")
plt.errorbar(pivot_mean_df.index,pivot_mean_df["Placebo"],yerr=pivot_sterr_df["Placebo"],color="black",marker="D",linewidth=0.5,elinewidth=1.5,label="Placebo")
plt.title("Tumor Response to Treatment")
plt.xlabel("Time (Days)")
plt.ylabel("Tumor Volume (mm3)")
plt.grid(axis="y")
plt.legend(loc="best")
# Save the Figure
plt.savefig("TumorResponseToTreatment.png")
```

```python
# Show the Figure
plt.show()
```

## Metastatic Response to Treatment

```python
# Store the Mean Met. Site Data Grouped by Drug and Timepoint 
grouped_df_3 = merge_data_df.groupby(['Drug', 'Timepoint'])
mean_metastatic = grouped_df['Metastatic Sites'].mean()
# Convert to DataFrame
mean_metastatic_df = pd.DataFrame(mean_metastatic)
# Preview DataFrame
mean_metastatic_df.head()
```

```python
# Store the Standard Error associated with Met. Sites Grouped by Drug and Timepoint 
grouped_df_3 = merge_data_df.groupby(['Drug','Timepoint'])
sterr_metastatic = grouped_df_3.sem()['Metastatic Sites']
# Convert to DataFrame
sterr_metastatic_df = pd.DataFrame(sterr_metastatic)
# Preview DataFrame
sterr_metastatic_df.head()
```

```python
# Minor Data Munging to Re-Format the Data Frames
pivot_mean_metastatic_df = mean_metastatic_df.pivot_table(index="Timepoint",columns="Drug")["Metastatic Sites"]

# Preview that Reformatting worked
pivot_mean_metastatic_df
```

```python
pivot_sterr_metastatic_df = sterr_metastatic_df.pivot_table(index="Timepoint",columns="Drug")["Metastatic Sites"]
pivot_sterr_metastatic_df
```

```python
# Generate the Plot (with Error Bars)
plt.errorbar(pivot_mean_metastatic_df.index,pivot_mean_metastatic_df["Capomulin"],yerr=pivot_sterr_metastatic_df["Capomulin"],color="red",marker="o",linewidth=0.5,elinewidth=1.5,label="Capomulin")
plt.errorbar(pivot_mean_metastatic_df.index,pivot_mean_metastatic_df["Infubinol"],yerr=pivot_sterr_metastatic_df["Infubinol"],color="blue",marker="^",linewidth=0.5,elinewidth=1.5,label="Infubinol")
plt.errorbar(pivot_mean_metastatic_df.index,pivot_mean_metastatic_df["Ketapril"],yerr=pivot_sterr_metastatic_df["Ketapril"],color="green",marker="s",linewidth=0.5,elinewidth=1.5,label="Ketapril")
plt.errorbar(pivot_mean_metastatic_df.index,pivot_mean_metastatic_df["Placebo"],yerr=pivot_sterr_metastatic_df["Placebo"],color="black",marker="D",linewidth=0.5,elinewidth=1.5,label="Placebo")
plt.title("Metastatic Spread During Treatment")
plt.xlabel("Time (Days)")
plt.ylabel("Met. Sites")
plt.grid(axis="y")
plt.legend(loc="best")
# Save the Figure
plt.savefig("MetastaticSpreadDuringTreatment.png")
# Save the Figure

# Show the Figure
plt.show()
```

## Survival Rates

```python
# Store the Count of Mice Grouped by Drug and Timepoint (W can pass any metric)
grouped_df_4 = merge_data_df.groupby(['Drug','Timepoint'])
mice_count = grouped_df_4.count()['Mouse ID']
# Convert to DataFrame
mice_count_df = pd.DataFrame(mice_count)
# Preview DataFrame
mice_count_df.head()
```

```python
# Minor Data Munging to Re-Format the Data Frames
pivot_mice_count_df = mice_count_df.pivot_table(index="Timepoint",columns="Drug")["Mouse ID"]
# Preview the Data Frame
pivot_mice_count_df
```

```python
# Generate the Plot (Accounting for percentages)
plt.errorbar(pivot_mice_count_df.index,pivot_mice_count_df["Capomulin"]/25 * 100,color="red",marker="o",linewidth=0.5,elinewidth=1.5,label="Capomulin")
plt.errorbar(pivot_mice_count_df.index,pivot_mice_count_df["Infubinol"]/25 * 100,color="blue",marker="^",linewidth=0.5,elinewidth=1.5,label="Infubinol")
plt.errorbar(pivot_mice_count_df.index,pivot_mice_count_df["Ketapril"]/25 * 100,color="green",marker="s",linewidth=0.5,elinewidth=1.5,label="Ketapril")
plt.errorbar(pivot_mice_count_df.index,pivot_mice_count_df["Placebo"]/25 * 100,color="black",marker="D",linewidth=0.5,elinewidth=1.5,label="Placebo")
plt.title("Survival During Treatment")
plt.xlabel("Time (Days)")
plt.ylabel("Survival Rate(%)")
plt.grid(axis="y")
plt.legend(loc="best")
# Save the Figure
plt.savefig("SurvivalDuringTreatment.png")
# Show the Figure
plt.show()
```
